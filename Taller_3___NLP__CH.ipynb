```
import bs4 as bs 
import urllib.request
```

```
enlace2 = ["https://es.wikipedia.org/wiki/Canis_familiaris",
           "https://es.wikipedia.org/wiki/Felis_silvestris_catus",
           "https://es.wikipedia.org/wiki/Psittacoidea",
           "https://es.wikipedia.org/wiki/Gorilla",
           "https://es.wikipedia.org/wiki/Lutrinae",
           "https://es.wikipedia.org/wiki/Panthera_leo",
           "https://es.wikipedia.org/wiki/Elephantidae",
           "https://es.wikipedia.org/wiki/Rhopalocera",
           "https://es.wikipedia.org/wiki/Ursidae",
           "https://es.wikipedia.org/wiki/%C3%81guila"]
```

```
import urllib.request
```

```
for x in range(len(enlace2)):
    request = urllib.request.Request(enlace2[x], headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    Links = bs.BeautifulSoup(source, 'html.parser')
    LinksHeader=Links.find("h1").text
    negrilla = Links.find("p").find_all("b")
    for y in range(len(negrilla)):
        negrilla[y]=negrilla[y].text
    print(LinksHeader)
    print(negrilla)
```

```
negrilla = Links.find("p").find_all("b")

for x in range(len(negrilla)):
  negrilla[x]=str(negrilla[x]).replace("<b>", "").replace("</b>", "")

negrilla
```
